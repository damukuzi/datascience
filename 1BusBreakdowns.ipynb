### Next Steps: Data Cleaning Steps


```
# Lets first start by creating a copy of our dataframe 
# df_clean = df.copy(). We will use this copy as our cleaning copy.
# ---
#
BusBreakdowns_clean = BusBreakdowns.copy()
BusBreakdowns_clean.head()
```

```
# Then extracting the lowest delay time in the column how_long_delayed from the string
# 
BusBreakdowns_clean['How_Long_Delayed'] = BusBreakdowns_clean['How_Long_Delayed'].str.extract('(\d+)')
BusBreakdowns_clean['How_Long_Delayed'].head()
```

We impute the null values in 'how_long_delayed' column with mean of the column. This will take a couple of steps...

```
# We first convert our how_long_delayed to float type to allow for imputation
BusBreakdowns_clean['How_Long_Delayed'] = pd.to_numeric(BusBreakdowns_clean['How_Long_Delayed'],downcast = "float")
BusBreakdowns_clean['How_Long_Delayed'].head()
```

```
# Then later perform our mean imputation 
# ---
BusBreakdowns_clean['How_Long_Delayed'] = BusBreakdowns_clean['How_Long_Delayed'].fillna(BusBreakdowns_clean['How_Long_Delayed'].mean())
BusBreakdowns_clean['How_Long_Delayed'].head()
```

```
# Then convert back our how_long_delayed column to integer datatype
BusBreakdowns_clean['How_Long_Delayed'] = BusBreakdowns_clean['How_Long_Delayed'].astype(int)
BusBreakdowns_clean['How_Long_Delayed'].head()
```

```
# Then check for nulls in the column
BusBreakdowns_clean.isnull().any()
```

```
# Rename Boro column to Borough
BusBreakdowns_clean =  BusBreakdowns_clean.rename(columns = {'Boro':'Borough'})
BusBreakdowns_clean.columns
```

```
# Lastly we convert all values in our colums to lower case
# for ease of reading
BusBreakdowns_clean.columns = BusBreakdowns_clean.columns.str.lower()
BusBreakdowns_clean.columns
```

```
# Check the first 5 record the cleaned dataset
# ---
#
BusBreakdowns_clean.head()
```


## 5. Solution Implementation

Here we investigate the questions that would help craft our recommendations.

### 5.a) Questions

```
# 1. Which bus companies that had the highest breakdowns?
# ---
#
breakdowns = BusBreakdowns_clean.groupby(['bus_company_name']).count()

# Sort to get bus company with highest breakdowns
# ---
breakdowns.sort_values(by='school_year',ascending=False)
```

```
# 2. What were the top 3 reasons for bus delays?
# ---
#
bus_delay = BusBreakdowns_clean.groupby(['reason']).count()[['how_long_delayed']].reset_index()

# Sort to get the most frequent reason
# ---
bus_delay.sort_values(by='how_long_delayed',ascending=False)
```

```
# 3. How many students were in the buses when they broke down? 
# ---
#
no_of_students = BusBreakdowns_clean.groupby(['number_of_students_on_the_bus']).count()[['busbreakdown_id']].reset_index()

no_of_students.head()
```

```
# 4. Which were most frequent reasons for bus breakdowns?
# ---
#
breakdown_reasons = BusBreakdowns_clean[BusBreakdowns_clean.breakdown_or_running_late == 'breakdown'].groupby(['reason']).count()

# Sort to get most frequent reasons
# ---
# YOUR CODE GOES BELOW
breakdown_reasons.head()
```

```
# 5. What were the most frequent reasons for the bus running late? 
# ---
#
reasons = BusBreakdowns_clean.groupby(['reason',]).count()

# Get the records with running late reasons and sort to get most frequent reasons
# ---
reasons.sort_values(by='how_long_delayed',ascending=False)
```

```
# 6. What was the average delay time for each reason type?
# ---
# 
avg_delay = BusBreakdowns_clean.groupby('reason').mean().reset_index()

# Get the records with reasons and how long on average a delay took then sort 
# ---
# YOUR CODE GOES BELOW
avg_delay.sort_values(by='how_long_delayed',ascending=False)
```
