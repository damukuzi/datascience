```
def CustomerBill(customer_usage):
    """ function receives a dictionary containing minutes and texts used by a customer
         and prints out the customer bill returns none"""
    minutes_used = customer_usage['minutes_used']
    texts_used = customer_usage['texts_used']

    basecharge = 10
    additionalminutesCharge = 0
    additionaltextsCharge = 0

    if minutes_used > 200:
        additionalminutesCharge = 0.03 * (minutes_used - 200)

    if texts_used > 200:
        additionaltextsCharge = 0.001 * (texts_used - 200)

    totalcharge = basecharge + additionalminutesCharge + additionaltextsCharge
    taxdue = round((totalcharge * 0.14),2)
    totalbill = round((totalcharge + taxdue),2)

    print ("your basecharge is {} extraminutes {} extratext {} taxdue {} and totalbill {} ".format(basecharge,additionalminutesCharge,additionaltextsCharge,taxdue,totalbill))

usage = {'minutes_used':400,'texts_used':300}
CustomerBill(usage)
```
