```
# getting the records with outliers
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

outliers_df = df[((df < (q1 - 1.5 * iqr)) | (df > (q3 + 1.5 * iqr))).any(axis = 1)]
print(outliers_df.shape)
outliers_df.sample(3)
```

```
# calculating percentage of outliers
round((outliers_df.shape[0] / df.shape[0]) * 100, 2)
```

```
# removing the records with the outliers
df = df[~((df < (q1 - 1.5 * iqr)) | (df > (q3 + 1.5 * iqr))).any(axis = 1)]
df.shape
```

Feature construction

```
# calculating the popularity per artist

# grouping by artist and calculating the average popularity
avg_artist_popularity_df = pd.DataFrame(df.groupby('track_artist')['track_popularity'].mean().reset_index().rename(columns={'track_popularity':'Avg artist popularity'}))

# adding the new column to our dataset
df = pd.merge(df, avg_artist_popularity_df, on='track_artist')
# previewing our modified dataset
df.head()
```

```
# calculating the popularity per track

# grouping by album and calculating the average popularity
avg_track_popularity_df = pd.DataFrame(df.groupby('track_id')['track_popularity'].mean().reset_index().rename(columns={'track_popularity':'Avg track popularity'}))

# adding the new column to our dataset
df = pd.merge(df, avg_track_popularity_df, on='track_id')
# previewing our modified dataset
df.sample(5)
```

```
# calculating the popularity per album

# grouping by album and calculating the average popularity
avg_album_popularity_df = pd.DataFrame(df.groupby('track_album_id')['track_popularity'].mean().reset_index().rename(columns={'track_popularity':'Avg album popularity'}))

# adding the new column to our dataset
df = pd.merge(df, avg_album_popularity_df, on='track_album_id')
# previewing our modified dataset
df.sample(5)
```

```
df.dtypes
```

```
# encoding 'room_type'
df['playlist_genre'] = df['playlist_genre'].replace({'pop' : '0', 'rap' : '1','rock' : '2','latin' : '3','r&b':'4','edm':'5'}).astype(int)
```

```
#dropping objects
cat_var = df.select_dtypes(include = ['object']).columns.to_list()
df.drop(columns = cat_var, inplace = True)
```

```
df.columns
```

```
# split into features (X) and label (Y) while removing categorical variables
X = df.iloc[:,[1,2,3,4,5,7,8,9,10,11,12,13,14,15]].values  
y = df.iloc[:, 0].values
```

```
# split into 70-30 train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state = 42)
```

```
# scaling our features
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler() 
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```
# fitting PCA to our data 
from sklearn.decomposition import PCA
pca = PCA(random_state = 0)
pca.fit(X_train)

# getting the explained variance ratio
evr = pca.explained_variance_ratio_
range_length = len(evr) + 1

# plotting the explained variance ratio so we can determine how any principal components to keep
plt.figure(figsize = (10, 8))
plt.plot(range(1, range_length), evr.cumsum(), marker = 'x', linestyle = '--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```
# fitting PCA with 8 components onto our features and transforming them
from sklearn.decomposition import PCA
pca = PCA(random_state = 0,n_components=8)
pca.fit(X_train)
```

```
# loading some regressors
from sklearn.svm import SVR 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor  

# setting up the regressors
svm_regressor = SVR(kernel='rbf', C=10)
knn_regressor = KNeighborsRegressor()
dec_regressor = DecisionTreeRegressor(random_state=27)

# fitting our regressors to our training data
svm_regressor.fit(X_train, y_train)
knn_regressor.fit(X_train, y_train)
dec_regressor.fit(X_train, y_train)

# making predictions
svm_y_pred = svm_regressor.predict(X_test)
knn_y_pred = knn_regressor.predict(X_test)
dec_y_pred = dec_regressor.predict(X_test)

from sklearn import metrics
from sklearn.metrics import r2_score
# evaluating RMSE and R2 scores
print('SVM RMSE:', np.sqrt(metrics.mean_squared_error(y_test, svm_y_pred)))
print('KNN RMSE:', np.sqrt(metrics.mean_squared_error(y_test, knn_y_pred)))
print('Decision Tree RMSE:', np.sqrt(metrics.mean_squared_error(y_test, dec_y_pred)))
```

segmentation

```
X = df.iloc[:,].values
```

```
# scale our features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# fit and transform our features
norm = sc.fit(X)
X = norm.transform(X)
```

```
# fitting KMeans with 5 clusters
from sklearn.cluster import KMeans

clusterer = KMeans(6, random_state=0)
clusterer.fit(X)
# getting the cluster labels
clusterer.labels_
```

```
# getting the KMeans scatter plot
kmeans_pred = clusterer.labels_
clusterer_pred = clusterer.predict(X)
kmeans_clust = np.unique(clusterer_pred)

plt.figure(figsize = (8, 6))
for cluster in kmeans_clust:
    row_ix = np.where(clusterer_pred == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label = cluster)
    
plt.title("Clustering by K-Means Clustering (k = 5)")
plt.legend()
plt.show()
```

```
# getting the optimum value of k for our KMeans model using the Elbow Method
Error = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(X)
    kmeans.fit(X)
    Error.append(kmeans.inertia_)

plt.plot(range(1, 11), Error, 'bx-')
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.grid(True);
```

```
new_clusterer = KMeans(4, random_state=0)
new_clusterer.fit(X)

kmeans_pred = new_clusterer.labels_
clusterer_pred = new_clusterer.predict(X)
kmeans_clust = np.unique(clusterer_pred)

plt.figure(figsize = (8, 6))
for cluster in kmeans_clust:
    row_ix = np.where(clusterer_pred == cluster)
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label = cluster)
    
plt.title("Clustering by K-Means Clustering (k = 5)")
plt.legend()
plt.show()
```

```
# add these cluster labels to our dataset
df['kmeans_cluster_group'] = new_clusterer.predict(X)

# previewing a sample of our dataset
df.head()
```
