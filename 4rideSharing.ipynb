```
import pandas as pd
```

```
traffic_accidents = pd.read_csv('traffic_accidents.csv')

traffic_accidents.head()
```

```
%sql DROP TABLE if EXISTS traffic_accidents;

%sql PERSIST traffic_accidents;
```

```
%sql SELECT * FROM traffic_accidents LIMIT 5;
```

```
%sql select make,count(*) from traffic_accidents group by make order by count(*) desc limit 2;
```

```
%sql select travel_direction,count(*) from traffic_accidents group by travel_direction order by count(*) desc limit 2;
```

```
%sql select count(*) from traffic_accidents where vehicle_use = 'PERSONAL';
```

```
%sql select count(*) from traffic_accidents where vehicle_use = 'PERSONAL' AND vehicle_type = 'SPORT UTILITY VEHICLE (SUV)';
```

```
%sql select count(*) from traffic_accidents where unit_type = 'PARKED' and make = 'MERCEDES-BENZ' ;
```

```
%sql select sum(occupant_cnt) from traffic_accidents;
```

```
%sql select travel_direction,sum(occupant_cnt) from traffic_accidents group by travel_direction order by sum(occupant_cnt) desc limit 2;
```

```
%sql select maneuver,make,count(*) from traffic_accidents where vehicle_defect = 'NONE' order by make limit 5;
```
